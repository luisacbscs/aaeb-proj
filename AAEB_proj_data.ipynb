```python
# read text file into pandas DataFrame
file1 = pd.read_table(path+r"\User_0EA27ICBLF.txt", delimiter=":",header=None)
print(file1)

indexes = []
for name in os.listdir(path):
    name = name.replace("User_","")
    name = name.replace(".txt","")
    indexes.append(name)

attributes = [att for att in file1[0]]
  
# display DataFrame
values = [value for value in file1[1]]
print(values)
```

```python
import glob
  
file_list = glob.glob(path + "/*.txt")

df = pd.DataFrame(columns=attributes, index=indexes)

for n in range(len(file_list)):
    data = pd.read_table(file_list[n], header=None, sep=":")
    values = [value for value in data[1]]
    df.loc[indexes[n]] = values

df.head(20)
```

```python
#Check if there are duplicate values
bool_dup = df.duplicated()

dup = 0
for el in bool_dup:
    if el == True:
        dup += 1
        
print('There are '+str(dup)+' duplicate values.')
```

```python
df.drop_duplicates(inplace=True)
```
